```python
def categorical_features(df):
    categorical_cols = df.select_dtypes(include=["object", np.number]).columns.tolist()
    types = df[categorical_cols].dtypes
    uniques = df[categorical_cols].nunique()
    categorical_tbl = pd.concat([uniques, types], axis=1).sort_values(0, ascending=False).rename(columns={0: 'Unique Values', 1: 'Data Types'})
    return categorical_tbl
```

```python
train = application_train.copy()
```

```python
train = train.sample(10000, random_state=50)
train
```

```python
train_labels = train.pop('TARGET')
```

```python
def encodeLabel(df):
    label_encoder = LabelEncoder()
    categorical_cols = df.select_dtypes(include=["object"]).columns.tolist()
    for col in categorical_cols:
        df[col] = label_encoder.fit_transform(df[col])

    return df
```

```python
train = encodeLabel(train)
```

```python
# imputer = SimpleImputer(missing_values=np.nan,strategy='median')
imputer=IterativeImputer(verbose=2)
scaler = MinMaxScaler(feature_range= (0,1))
```

```python
imputer.fit(train)
train = imputer.transform(train)
```

```python
X_train, X_test, y_train, y_test = train_test_split(train, train_labels, test_size=0.2, random_state=101)
```

Ramdom Forest Regression Model

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor(n_estimators=100,random_state=0,verbose=5)
```

```python
# Training the Random Forest model of Train Data Sets
rfr.fit(X_train,y_train)
```

```python
# Predictiing the result of the Random Forest model
y_pred = rfr.predict(X_test)
```

```python
# Accuracy Score of the Random Forest Model
print('Accuracy Score: {}'.format(metrics.accuracy_score(y_test, y_pred.round())))
```

```python
# ROC AUC Score of the model
print('ROC AUC Score: {}'.format(metrics.roc_auc_score(y_test,y_pred)))
```

```python
false_positive_rate1, true_positive_rate1, threshold1 = roc_curve(y_test, y_pred)
```

```python
plt.subplots(1, figsize=(10, 10))
plt.title('Receiver Operating Characteristic - Random Forest Iterative')
plt.plot(false_positive_rate1, true_positive_rate1)
plt.plot([0, 1], ls="--")
plt.plot([0, 0], [1, 0], c=".7"), plt.plot([1, 1], c=".7")
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

```python
from sklearn.model_selection import cross_val_score
n_folds = 10
scores = cross_val_score(rfr, X_train, y_train, cv=n_folds, scoring='roc_auc', n_jobs=-1)
fold_names = list(range(n_folds))
fold_names.append('Average')

avg_score = scores.mean()
scores = list(scores)
scores.append(avg_score)
cv_score = pd.DataFrame({'Fold Index': fold_names, 'ROC AUC (Norm)': scores, })

cv_score
```
